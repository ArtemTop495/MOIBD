```python
# GET - –∑–∞–ø—Ä–æ—Å
url = 'https://stopgame.ru/games/pc/best' # —Å—Ç—Ä–∞–Ω–∏—Ü–∞ —Å–æ –≤—Å–µ–º–∏ —Å—Ç–∞—Ç—å—è–º–∏ 
page = requests.get(url)
page.encoding = 'utf-8'
```

```python
result_list = {'title': [], 'rating': [], 'date': [], 'tags': []}
```

### –ê–ª–≥–æ—Ä–∏—Ç–º

```python
for pagenum in range(1, 6):
    url = 'https://stopgame.ru/games/pc/best' + '?p=' + f'{pagenum}'
    print(url)
    page = requests.get(url)
    soup = bs(page.text, 'html.parser')
    titles = soup.find_all('a', class_='_card_32u6p_4')
    
    for i in range(0, len(titles)):
        print(titles[i].get('title'))
        result_list['title'].append(titles[i].get('title'))
        print(titles[i].text[-6:-2] + '/ 5.0')
        result_list['rating'].append(titles[i].text[-6:-2] + '/ 5.0')
        game_url = 'https://stopgame.ru' + titles[i].get('href')
        print(game_url)
        game_page = requests.get(game_url)
        game_soup = bs(game_page.text, 'html.parser')
        date = game_soup.find('dl', class_='_game-info__grid_fdfpg_989').find('dd')
        print(date.text.replace('<dd>','').replace('</dd>',''))
        result_list['date'].append(date.text.replace('<dd>','').replace('</dd>',''))
        tags_list = game_soup.find_all('a', class_='_tag_n52si_624')
        tags = ''
        for i in tags_list:
            tags += i.text + ", "
        tags = tags[0:-2]
        print(tags+ "\n")
        result_list['tags'].append(tags)
```

```python
print(len(result_list['title']))
print(len(result_list['rating']))
print(len(result_list['date']))
print(len(result_list['tags']))
```

```python
file_name = 'games.csv'
df = pd.DataFrame(data=result_list)
df.to_csv(file_name)
df.head(100)
```

```python
def extract_text_from_pdf(pdf):
    doc = pymupdf.open(pdf)
    text = ""
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text += "\n" + page_text
    return text
```

### –ê–ª–≥–æ—Ä–∏—Ç–º

```python
all_names = []
all_descriptions = []
all_hubs = []
all_raitings = []
all_subscribers = []
def pars_pdf(text):
    text2 = text.split('\n')
    print(text2)
    for i in range(len(text2)-1, 0, -1):
        if text2[i] in ['–í—Å–µ –ø–æ—Ç–æ–∫–∏', '–î–æ–±–∞–≤–∏—Ç—å –∫–æ–º–ø–∞–Ω–∏—é', '–ù–û–í–û–°–¢–ò', '+50','–°–¢–ê–¢–¨–ò +64','–°–¢–ê–¢–¨–ò', '+64', '–°–¢–ê–¢–¨–ò +63','–ù–û–í–û–°–¢–ò +51','–•–ê–ë–´','–ê–í–¢–û–†–´','–ö–û–ú–ü–ê–ù–ò–ò','–ü–æ–∏—Å–∫','–ù–∞–∑–≤–∞–Ω–∏–µ ','–†–µ–π—Ç–∏–Ω–≥ ','–ü–æ–¥–ø–∏—Å—á–∏–∫–∏','–ö–ê–ö –°–¢–ê–¢–¨ –ê–í–¢–û–†–û–ú','–ò—Å—Å–ª–µ–¥—É–µ–º –î—É—Ö–æ–≤ –ú–∞—à–∏–Ω –≤ –Ω–æ–≤–æ–º –°–µ–∑–æ–Ω–µ –Ω–∞ –•–∞–±—Ä–µ'] or text2[i] == '':
            text2.pop(i)
    for i in range(len(text2)-1, 0, -1):
        if text2[i].__contains__('–ü–∏—à–µ—Ç –≤ —Ö–∞–±—ã:'):
            hubs = text2[i]
            if (text2[i+1].isnumeric() or text2[i+1].__contains__(".")) and (text2[i+2].__contains__("K") or text2[i+2].isdecimal()):
                raiting = text2[i+1]
                subscribers = text2[i+2]
            elif (text2[i+2].isnumeric() or text2[i+2].__contains__(".")) and (text2[i+3].__contains__("K") or text2[i+3].isdecimal()):
                hubs += text2[i+1]
                raiting = text2[i+2]
                subscribers = text2[i+3]
            else:
                for j in text2[i+2]:
                    if (j.isdigit()):
                        hubs += text2[i+2][0:text2[i+2].index(j)-1]
                        raiting = text2[i+2][text2[i+2].index(j):-1]
                        subscribers = text2[i+3]
            description = text2[i-1]
            name = text2[i-2]
            print(f"\n\n\n{name}\n{description}\n{hubs}\n{raiting}\n{subscribers}\n\n\n")
            text2.pop(i)
            text2.pop(i+1)
            text2.pop(i+2)
            if not ((text2[i+1].isnumeric() and (text2[i+2].__contains__("K") or text2[i+2].isnumeric())) and (text2[i+2].isnumeric() and (text2[i+3].__contains__("K") or text2[i+3].isnumeric()))):
                text2.pop(i+3)
            text2.pop(i-1)
            text2.pop(i-2)                    
            all_names.append(name)
            all_descriptions.append(description)
            all_hubs.append(hubs)
            all_raitings.append(raiting)
            all_subscribers.append(subscribers)
pars_pdf(extract_text_from_pdf(r"PDF2\–ö–æ–º–ø–∞–Ω–∏–∏ _ –•–∞–±—Ä.pdf"))
pars_pdf(extract_text_from_pdf(r"PDF2\–ö–æ–º–ø–∞–Ω–∏–∏.pdf"))
```

### –°–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞

```python
df = pd.DataFrame({
'NameCompany': all_names,
'Description': all_descriptions,
'Hubs': all_hubs,
'Raiting': all_raitings,
'Subscribers': all_subscribers
})
df
```

# –ü–∞—Ä—Å–∏–Ω–≥ json —Ñ–∞–π–ª–∞

#### –ü–æ–ª—É—á–µ–Ω–∏–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ —Å —Ñ–∞–π–ª–∞

```python
with open(r'Data\GooseGaming.json', 'r', encoding="utf8") as file:
    data = json.load(file)
```

```python
all_ref_text = []
all_ref_day = []
all_ref_month = []
all_ref_time = []
all_ref_rate = []
all_ref_views = []
def pars_json(local_data):
    for i in local_data['refs']:
        all_ref_text.append(i[0])
        all_ref_day.append(i[1]['day'])
        all_ref_month.append(i[1]['month'])
        all_ref_time.append(i[1]['time'])
        all_ref_rate.append(i[2]['rate'])
        all_ref_views.append(i[2]['views'])
pars_json(data)
```

```python
df = pd.DataFrame({
    'ref text': all_ref_text,
    'ref day': all_ref_day,
    'ref month': all_ref_month,
    'ref time': all_ref_time,
    'ref rate': all_ref_rate,
    'ref views': all_ref_views
})
df
```
